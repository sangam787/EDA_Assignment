# Que-3 What is the distribution of car prices for each transmission type, and how do the interquartile ranges compare?
#**Ans-3**

```python
# Visulizing the distribution using box plots

plt.figure(figsize=(8, 6))
sns.boxplot(x= 'Transmission', y='Price ($)', data=df_copy)
plt.title('Distribution of Car Prices for Each Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Price ($)')
plt.show()
```

```python
# Calculating and compare the interquartile ranges

def iqr(data):
  Q1 = data.quantile(0.25)
  Q3 = data.quantile(0.75)
  return Q3 - Q1

transmission_types = df_copy['Transmission'].unique()
for transmission_type in transmission_types:
  data = df_copy[df_copy['Transmission'] == transmission_type]['Price ($)']
  IQR = iqr(data)
  print(f"IQR for {transmission_type}: {IQR}")
```

# Que-12           Is there any seasonal trend in car sales based on the date of sale?
#**Ans-12**

```python
# converting the 'Date' column to datetime
df_copy['Date'] = pd.to_datetime(df_copy['Date'])


# Assuming 'Date' column is in datetime format
df_copy['Month'] = df_copy['Date'].dt.month
# alternatively, extract quarter
df_copy['Quarter'] = df_copy['Date'].dt.quarter

# grouping by mont and calculating total sales
monthly_sales = df_copy.groupby('Month')['Price ($)'].sum()
# grouping by quarter and calculate total sales
quarterly_sales = df_copy.groupby('Quarter')['Price ($)'].sum()

# visualizing the trend:
print("Line plot for monthly sales:")

plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='line', marker= 'o')
plt.title('Monthly Car Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.tight_layout()
plt.show()

print("Bar plot for quarterly sales:")
plt.figure(figsize= (8, 6))
quarterly_sales.plot(kind='bar')
plt.title('Quarterly Car Sales Trend')
plt.xlabel('Quarter')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

# Que-6 What are the most popular product keywords?

```python
import collections

# combining product names into a single string
all_product_names = ' '.join(df_copy['product_name'].astype(str).tolist())

# spliting into words
words = all_product_names.lower().split()

# counting word frequency
word_counts = collections.Counter(words)

# getting the most frequent words (keywords)
most_common_keywords = word_counts.most_common(20)

# filtering out common words (stop words)
stop_words = ["for", "the", "and", "with", "in", "of", "to", "a", "is", "on", "&","it", "this", "that", "by", "are", "be", "as", "you", "your", "-", "(", ")", "[", "]", "{", "}", "/", "|", ",", ".", ";", ":", "'", '"']

popular_keywords = [(word, count) for word, count in most_common_keywords if word not in stop_words]
print(popular_keywords)
```

#Que-6 Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and display the average popularity for each artist.
#**Ans-6**

```python
# finding the top 5 most popular artist:
top_5_artists = df_copy['Artist'].value_counts().head(5).index

# calculating average popularity for each artist:
average_popularity= df_copy[df_copy['Artist'].isin(top_5_artists)].groupby('Artist')['Popularity'].mean()

print("Average Popularity for Top 5 Artists:")
print(average_popularity)

artist_with_highest_avg_popularity = average_popularity.idxmax()
print(f"\nArtist with the highest average popularity: {artist_with_highest_avg_popularity}")
```
